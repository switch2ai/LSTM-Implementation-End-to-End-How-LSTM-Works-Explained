```python
## Components of LSTM :
State :
-- Hidden state : It stores short term information
-- Cell state : It stores long term information (imp long term info)

Gates : Used to manage flow of information in cell state
-- Forget Gate : Remove irrelevent information from cell state
-- Input Gate :  Add new relevent information to cell state
-- Output Gate : output of LSTM cell
```
